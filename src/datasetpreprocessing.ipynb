```python
import os
import shutil

# Paths
base_dir = r'C:\Users\acess\OneDrive\Documents\Python Projects\Recipe_miniproject\data'
output_dir = r'C:\Users\acess\OneDrive\Documents\Python Projects\Recipe_miniproject\splitted_data'

def split_dataset(base_dir, output_dir, split_ratio=(0.7, 0.2, 0.1)):
    if not os.path.exists(base_dir):
        raise FileNotFoundError(f"The directory {base_dir} does not exist.")
    
    classes = os.listdir(base_dir)
    if not classes:
        raise ValueError(f"No classes found in the directory {base_dir}. Ensure it contains subfolders for each recipe class.")
    
    for cls in classes:
        cls_path = os.path.join(base_dir, cls)
        if not os.path.isdir(cls_path):
            print(f"Skipping {cls_path} as it is not a directory.")
            continue
        
        files = os.listdir(cls_path)
        if not files:
            print(f"Skipping {cls_path} as it contains no files.")
            continue

        # Shuffle and split files
        split1 = int(len(files) * split_ratio[0])
        split2 = split1 + int(len(files) * split_ratio[1])
        splits = {
            'train': files[:split1],
            'val': files[split1:split2],
            'test': files[split2:]
        }
        
        for split, split_files in splits.items():
            split_dir = os.path.join(output_dir, split, cls)
            os.makedirs(split_dir, exist_ok=True)
            for file in split_files:
                shutil.copy(os.path.join(cls_path, file), os.path.join(split_dir, file))

# Call function to split the dataset
split_dataset(base_dir, output_dir)
print(f"Dataset successfully split into train, val, and test sets at {output_dir}.")
```
