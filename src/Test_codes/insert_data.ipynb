```python
import os

def print_directory_structure(folder_path, indent=0, skip_folders=None):
    if skip_folders is None:
        skip_folders = []  # Default to an empty list if no folders to skip

    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        
        # Skip the specified folders
        if item in skip_folders and os.path.isdir(item_path):
            continue
        
        print("  " * indent + "|-- " + item)  # Print the folder structure
        
        # Recursively print subdirectory contents
        if os.path.isdir(item_path):
            print_directory_structure(item_path, indent + 1, skip_folders)

# Specify the folder path
folder_path = r"C:\Users\acess\OneDrive\Documents\Python Projects\Recipe_miniproject"
skip_folders = ["data", "splitted_data"]  # Folders to skip
print_directory_structure(folder_path, skip_folders=skip_folders)
```
