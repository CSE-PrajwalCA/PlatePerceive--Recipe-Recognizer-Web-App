```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model

# Set the path for the test dataset and the saved .keras model file
model_path = r'C:\Users\acess\OneDrive\Documents\Python Projects\Recipe_miniproject\src\vgg19_recipe_classifier.keras'  # Path to your .keras file
test_dir = r'C:\Users\acess\OneDrive\Documents\Python Projects\Recipe_miniproject\splitted_data\test'  # Path to your test data

# If the model has custom layers, pass them as custom_objects
# For example, if your model has a custom layer, include it here like this:
# model = load_model(model_path, custom_objects={'MyCustomLayer': MyCustomLayer})

# Step 1: Load the pre-trained model
try:
    model = load_model(model_path)
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

# Step 2: Prepare the test dataset (same preprocessing as training)
test_datagen = ImageDataGenerator(rescale=1.0 / 255)  # Rescale the images to [0, 1]

# Step 3: Create the test data generator
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),  # Adjust the size based on your model's input size
    batch_size=32,
    class_mode='categorical',  # Assuming multi-class classification
    shuffle=False  # No shuffling for evaluation, as we want to evaluate on a fixed set
)

# Step 4: Evaluate the model on the test dataset
loss, accuracy = model.evaluate(test_gen)

# Print the results
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
